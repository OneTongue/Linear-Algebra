##### Plotting Multiple Vectors on 2D

```python
# Importing Libraries
import numpy as np
import matplotlib.pyplot as plt 
```

```python
# Origin
X = [0, 0, 0]
Y = [0, 0, 0]

# Vector Direction
U = [2, -3, -1]
V = [-1, 4, 3]
```

```python
# Plotting the vectors using pyplot.quiver()
plt.quiver(
    X,  # Array for the starting points in X of the vectors
    Y,  # Array for the starting points in Y of the vectors
    U,  # Array for the change in X of the Vectors
    V,  # Array for the change in Y of the Vectors
    angles="xy",
    scale_units="xy",
    scale=1,
    color=["r", "g", "b"],
    label=["Vector 1", "Vector 2"],
)

plt.plot(X, Y, "o", color="black", label="Start")

plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Plotting multiple vectors")
plt.grid(True)
plt.gca().set_aspect("equal", adjustable="box")

plt.show()
```
